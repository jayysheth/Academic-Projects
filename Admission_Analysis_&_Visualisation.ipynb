```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file into a DataFrame
file_path = "C:\\Users\\jaysh\\Machine Learning 1\\admission_data.csv"
admission_df = pd.read_csv ("admission_data.csv")
```

```python
# Display the first few rows of the dataset
print(admission_df.head())
```

```python
# Basic Analysis: Count the number of accepted and rejected admissions
admission_status_counts = admission_df['Admission_Status'].value_counts()
print("\nAdmission Status Counts:")
print(admission_status_counts)
```

```python
# Basic Analysis: Calculate the average GPA of accepted and rejected students
average_gpa = admission_df.groupby('Admission_Status')['GPA'].mean()
print("\nAverage GPA by Admission Status:")
print(average_gpa)
```

```python
# Visualization 1: Bar chart of admission status counts
plt.figure(figsize=(8, 6))
admission_status_counts.plot(kind='bar', color=['green', 'red'])
plt.title('Admission Status Counts')
plt.xlabel('Admission Status')
plt.ylabel('Number of Students')
plt.show()
```

```python
# Visualization 2: Box plot of GPA by admission status
plt.figure(figsize=(8, 6))
admission_df.boxplot(column='GPA', by='Admission_Status', grid=False, color=dict(boxes='blue', whiskers='blue', medians='red', caps='black'))
plt.title('GPA Distribution by Admission Status')
plt.suptitle('')  # Suppress the default title to avoid overlap
plt.xlabel('Admission Status')
plt.ylabel('GPA')
plt.show()
```
